자기상관 그래프처럼 무작위성을 파악하는데도 유용

=> 계절효과인 계정성이나 결정적인 추세가 있는지 감지하기 위해 사용한다.

```python
def get_autocorrelation_dataframe(series):
    def r(h):
        return ((data[:n - h] -mean) * (data[h:] - mean)).sum() / float(n) / c0
    
    n = len(series)
    data = np.asarray(series)
    mean = np.mean(data)
    c0 = np.sum((data-mean) ** 2) / float(n)
    x = np.arange(n) + 1
    y = map(r,x)
    df = pd.DataFrame(y, index=x)
    return df
```

### 평균회귀모델
 [가정]
    - 시계열 데이터는 과거의 평균값으로 회귀하려는 경향이 있다.
    - 어떤 변수가 다음과 같은 정규분포를 따른다고 하면 평균에 가까이 갈 확률이 높아지고, 반대로 멀어질수록 확률은 낮아지는것과 비슷한 이치라고 할수있따.
    
    즉, 현시점의 주가가 평균주가보다 낮으면 주가가 올라가려 하고, 반대로 현주가가 평균주가보다 높으면 주가가 내려가려 하는 경향을 이용한것.
    하지만, 실제 주가데이터는 무작위, 즉 랜덤워크라서 무턱대고 평균회귀모델을 적용할수없다
    
평균회귀 모델을 적용하기 위해, 적용하려는 주가 데이터가 랜덤워크인지 아닌지를 파악해야한다.
주가 데이터가 랜덤워크를 보인다는 것은 주가의 흐름이 이전의 데이터와 상관없이 결정됨을 의미하는데, 이는 평균회귀 모델을 적용할수없다,
평균회귀모델은 이전데이터가 현재데이터에 영향을 미쳐 평균으로 회귀하도록 해야하기 때문이다. 즉, 비독립적이여야 한다.

평균회귀 모델을 적용할수있는 시계일인지 아닌지를 다음과 같은 방법으로 판별해보다
    - ADF 테스트
    - 허스트 지수 

#### ADF 테스트

주가 데이터가 평균회귀 성질이 있다면, 현주가는 다음 주가가 어떻게 될지 예측가능 한 통계학적 정보를 포함한다.
예를들어, 현 주가가 평균보다 낮다면 다음주가의 값은 알수가 없디만, 주가가 위로 향하는 방향성을 갖을 확률이 높다는 것은 알수있다.

ADF 테스트는 어떤 시계열 데이터가 랜덤워크를 따른다는 가설을 세우고, 이 가설을 검증하여 랜덤워크인지 아닌지를 판단하는 방법이다

수식쓰기


ADF 테스트는 앞의 식에서 알파,베타,감마 = 0 을 대입해 가설검정을 실시한다. 만약 감마 = 0 이라는 가설 검정에서 실채하면 시계열 데이터는 랜덤워크가 아니라는 것을 증명하게 되고, 랜덤워크가 아니라면 평균회귀 시계열이라는 결론을 갖는다.

#### 허스트 지수
정상과정은 평균과 표준편차가 일정해서 표류하는 랜덤워크인 기하학 브라운 운동(GBM)보다 천천히 값이 확산된다.
주가 데이터는 시계열 데이터로, 시간에 따른 주식값의 변화를 확산이라고 생각할수있고, 이때 얼마나 빠르게 퍼져나가는지를 수학적 계산을 통해 측정할수있다.

허스트 지수의 핵심 아이디어는 분산을 확산속도로 치환가능하고, 그 값을 GBM의 확산속도와 비교하여 랜덤워크인지, 정상과정인지 파악할수있다는것이다.

$${Var(\tau) = [|log(t+\tau) - log(t)|^2] }$$ 
 
 GBM은 표류하는 랜덤워크 모델이어서 데이터가 충분히 크면, $ \tau $ 의 분산은 $ \tau $ 에 비례해야하므로 
 
  $${ [|log(t+\tau) - log(t)|^2] \sim \tau  }$$ 
  
이 식에서 사용한 ~ 은 충분히 큰 데이터라면 어떤 값으로 수렴할 것으로 의미다. 어떤 정상과정이 이 식을 만족하지 않는다면, 해당 정상과정은 평균회귀 성향이나 추세성향이 있다는 것을 알수있다.

$${ [|log(t+\tau) - log(t)|^2] \sim \tau^2~^H }$$ 

이 식에서 H를 허스트 지수라고 하고, H의 크기에 따라 해당 시계열 데이터의 확산속도가 어떤 형태를 띠게 될지 짐작가능하다.
어떤 시계열 데이터가 GBM이라 가정하면, $ \tau $의 지수가 2이어야 하므로 2H = 1 -> H = 0.5를 가져야 하고, GBM이 아니라면 H는 0또는 0에 가까운 값을 가져야한다. 이와 같이 계산한 허스트 지수의 값이 H < 0.5면 평균회귀, H > 0.5 면 발산하는 형태의 추세 성향이 있다는 것을 알수있다.

더 자세한 설명을 위해 허스트 지수를 정의한 식의 필요부분을 보자
$${ f(x) = \tau^2~^H }$$ 

이 함수에서 H에 각각 0, 0.5, 1 의 값을 넣어보면 다음과 같다.
    - H = 0.5, $${ f(x) = \tau^1 }$$
        H = 0.5 의 경우는 GBM인것을 알수있다.
    - H = 0, $${ f(x) = \tau^0 }$$
        H = 0 의 경우는 값이 1이 되므로 값이 평균회귀한다.
    - H = 1, $${ f(x) = \tau^2 }$$
        H = 1 의 경우는 지수함수가 되므로 발산하는 형태의 추세성향이 있음을 알수있다.
    
    허스트 지수는 정상과정이 평균회귀나 추세 성향이 있는지를 알수있을뿐 아니라 경향의 정도도 알수있는 매우 유용한 지수이다
    H가 0 에 가까우면 평균회귀성향이 강하고, 반대로 1에 가까울수록 추세성향이 강하다고 판단가능하다.
    
    
  

```python
def get_hurst_exponent(df):
    lags = range(2,100)
    ts = np.log(df)
    
    tau = [np.sqrt(np.std(np.subtract(ts[lag:],ts[:-lag]))) for lag in lags]
    poly = np.polyfit(np.log(lags), np.log(tau),1)
    
    result = poly[0]*2.0
    
    return result
```

```python
df_samsung = load_stock_data('samsung.data')
df_hanmi = load_stock_data('hanmi.data')

hurst_samsung = get_hurst_exponent(df_samsung['Close'])
hurst_hanmi = get_hurst_exponent(df_hanmi['Close'])
print("Hurst Exponent : Samsung = %s, Hanmi = %s" % (hurst_samsung,hurst_hanmi))
```

삼성전자 종가의 허스트지수와 한미약품의 종가의 허스트지수는 둘다 0.5에 가까우므로, GBM에 가깝다는 것을 알수있다.
즉 뚜렷한 평균회귀 성향이나 발산추세성향은 보이지 않는것을 알수있다.

 #### 평균회귀의 Half-life
 
 ADF테스트와 허스트 지수를 통해 삼성전자 주가와 한미약품 주가가 평균회귀모델을 적용하기에 부적합하다.
 주가 데이터 대부분이 평균회귀 모델적용 가능 테스트에서 실패한다. 하지만,2가지 테스트를 실했더라도 평균회귀 모델을 적용할수있는지를 알수있는 HALF-LIFE값을 알아보겠다.

HALF-LIFE는 값이 평균으로 회귀하는데 걸리는 시간을 의미한다. 이 값을 이용해 평균회귀모델을 적용할수있는 주가를 찾을수있다. 평균회귀성향이 있는 랜덤과정을 오른스타인-우렌벡 과정(ornstein-uhlenbeck process)라고 하며, 다음과 같은 확률미분방정식을 만족한다. 
 $${ dx_t = \lambda(\mu - x_t)dt + \sigma d W_t }$$ 
 $ \lambda $ 는 평균회귀 속도고, $ \mu $ 는 평균, $ dW_t $는 에러텀이다.
 
 HALF-LIFE를 쉽게 구하기 위해서 다음과 같은 오른스타인-우렌벡 과정을 가정해보자. $ y_0 $ 는 초기치, $ \lambda $는 평균회귀속도이다
 $${f(t) = y_0e^{-\lambda} t }$$
 
 이 프로세스는 평균회귀 하기 때문에 시간 t의 1/2에 대해서 
  $${f(t_{1/2}) = {f(t)/2} }$$ 
  이런 관계가 성립해야한다. 위의 2개의 식을 풀고 평균회귀를 위해 부호를 조정하면 다음과 같은 결과를 얻는다
  
  $${ Half-life, t_{1/2} = -{ln2/\lambda} }$$ 
  
  식을 통해 알수있듯, HALF-LIFE 는 평균회귀속도인 람다와 반비례관계에 있다. 람다가 커지면 속도가 빨라서 빨리 평균으로 회귀하고, 반대로 작아지면 회귀가 늦다는것을 의미한다.
  
  HALF-LIFE에서 의미하는 수치는 꼐산에 사용한 시간 단위이기 때문에 결과값도 단위시간에 따른 평균회귀 시간을 나타낸다는것을 유의해야한다. HALF-LIFE의 값이 크고 작음은 적용 전략에 따라 좋을수도, 나쁠수도 있다. 
  예를 들면 HALF-LIFE의 값이 크다는것은 장기간 지속하는 경향이 있다는것을 의미할수도 있고, 반대로 작다는것은 그만큼 주식값의 변동이 잦다는것으로 해석할수도 있기 때문에 전략에 맞춰 판단해야한다.

```python
def get_half_life(df):
    price = pd.Series(df)
    lagged_price = price.shift(1).fillna(method="bfill")
    delta = price - lagged_price
    beta = np.polyfit(lagged_price, delta, 1)[0]
    half_life = (-1*np.log(2)/beta)
    
    return half_life
```

```python
df_samsung = load_stock_data('samsung.data')
df_hanmi = load_stock_data('hanmi.data')

half_life_samsung = get_half_life(df_samsung['Close'])
half_life_hanmi = get_half_life(df_hanmi['Close'])
print("half_life : Samsung = %s, Hanmi = %s" % (half_life_samsung,half_life_hanmi))
```

삼성전자의 주가가 한미약품 주가보다 더 빠르게 평균으로 회귀함을 알수있다. 한미약품의 half-life값은 너무 커서 평균회귀 성향이 희박하다고 할수있다.

### 평균회귀 모델 구현
ADF 테스트, 허스트 지수, Half-life를 모두 통과한 종목을 찾았다면, 평균회귀모델을 구현해보자
평균회귀 모델의 기본개념은 주가가 평균보다 낮으면 주식을 매입하고, 반대로 평균보다 높으면 주식을 매도해 수익을 만드는 것이다. 

평균회귀 모델을 완성하려면 다음의 3가지 사항을 결정해야한다.
  - 평균 정의
  
    : 주가매도, 매수의 비교치로 사용될 평균을 어떻게 구할것인가?
    
    : 평균을 과거 특정기간의 주가로 구할 수도 있고, 아니면 이동평균을 이용해 계산할수도 있다.
    
    : 이동평균을 이용한다면, 이동평균의 기간 단위 또한 결정해야한다. 평균은 매도, 매수를 알려주는 일종의 신호로 활용한다.
         
  - 매도,매수 기준 
  
    : 예를 들어 주가가 평균보다 높아지는 순간에 매도할지 평균과 주가의 차이가 2배이상 될때 매도할지 결정하는 기준을 말한다.
                                
    : 평균과 마찬가지로 매도, 매수 기준은 모델의 수익률에 막대한 영향을 미치는 중요한 요소이다. 기준이 너무 낮다면 큰 이익을 보지 못하고, 반대로 너무 높다면 매도나 매수시기를 놓칠수있다.
    
  - 데이터 선택
  
     : 종가를 사용할지, 시작가를 사용할지 또는 현 시가를 사용할지 선택. 데이터 선택 역시 매우 중요한 문제로 간단히 선택하기에 무게가 가볍지않다.
     
     
     
     
 나는 다음과 같은 기준으로 평균회귀 모델을 구현한다.
 
   - 평균정의  : 10일 이동평균
 
   - 매도 매수기준 :  이동평균과 주가의 차이가 표준편차보다 크면 매도,매수 실행
      
      |주가 - 이동평균 | > 표준편차  일때, 주가 - 이동평균 > 0 : 매도   
                                   주가 - 이동평균 < 0 : 매수
    
   - 데이터 선택 : 종가 사용
